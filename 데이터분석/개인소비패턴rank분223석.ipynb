```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

path = 'C:/Users/zofld/Desktop'
os.listdir(path)
```

```python
personal_data = pd.read_csv(os.path.join(path, '머희소비내역.csv'), encoding='utf-8')
personal_data
```

```python
df = personal_data.pivot_table(index=['CLASS1'], 
                               values = ['카드이용건수계', '카드이용금액계'], aggfunc = 'sum')
personal_Df = df.reset_index()
personal_Df
```

```python
personal_Df['카드이용건수계순위']=personal_Df['카드이용건수계'].rank(method='min',ascending=False)
personal_Df['카드이용금액계순위']=personal_Df['카드이용금액계'].rank(method='min',ascending=False)
personal_Df
```

```python
df_list = []
for i in personal_Df['카드이용금액계순위']:
    df_list.append(i)
df_list
```

```python
data2 = pd.read_csv(os.path.join(path, '카드데이터 샘플.csv'), encoding='utf-8')
```

```python
data3 = data2.loc[3:16, :]
```

```python
from scipy import stats
```

```python
df_rank=data3['삼성카드'].rank(method='min',ascending=False)
```

```python
card_ranking=[]
for i in df_rank:
    card_ranking.append(i)
card_ranking
```

```python
tau, p_value = stats.kendalltau(df_list, card_ranking)
print(tau)
```

```python
final_card_ranking = [] ##유사도 저장
count = 1
for i in data3.iloc[:, 1:len(data3)] :
    df_real = data3.iloc[:, [count]]
    df_ranking=df_real.rank(method='min',ascending=False)
    card_rank=[]
    for j in df_ranking:
        card_rank.append(j)
    tau, p_value = stats.kendalltau(df_list, card_ranking) ##유사도 측정
    final_card_ranking.append(tau)
        
```

```python
final_card_ranking
```

```python
data_score = data2.loc[17, :]
```

```python
data_score
```

```python
len(final_card_ranking)
```

```python
score_index = 0
top_card_score = 0
for i in range(1,len(final_card_ranking)+1):
    if final_card_ranking[i-1] > 0.2 : #특정유사도 이상
        if int(data_score[i]) > top_card_score:
            top_card_score = int(data_score[i])
            score_index = i 
```

```python
real_card = data2.loc[0, :]
real_card[score_index]
print("카드명 : " + real_card[score_index] + ",  카드점수 : " + data_score[score_index])
```
