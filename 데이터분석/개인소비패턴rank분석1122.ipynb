```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

path = 'C:/Users/zofld/Desktop'
os.listdir(path)
```

```python
personal_data = pd.read_csv(os.path.join(path, '머희소비내역.csv'), encoding='utf-8')
```

```python
df = personal_data.pivot_table(index=['CLASS1'], 
                               values = ['카드이용건수계', '카드이용금액계'], aggfunc = 'sum')
personal_Df = df.reset_index()
```

```python
personal_Df['카드이용건수계순위']=personal_Df['카드이용건수계'].rank(method='min',ascending=False)
personal_Df['카드이용금액계순위']=personal_Df['카드이용금액계'].rank(method='min',ascending=False)
personal_Df
```

```python
##개인 소비내역 순위분석
df_list = []
for i in personal_Df['카드이용금액계순위']:
    df_list.append(i)
df_list
```

```python
##비교용 카드데이터
data2 = pd.read_csv(os.path.join(path, '카드데이터 샘플.csv'), encoding='utf-8')
```

```python
##카드데이터에서 업종부분만 추출
data3 = data2.loc[3:16, :]
data3
```

```python
##업종별 이름순 정렬
data3=data3.sort_values(by=data3.columns[0], ascending=True)
```

```python
data3=data3.reset_index()
data3=data3.drop(['index'],axis=1)
data3
```

```python
from scipy import stats
```

```python
data3.iloc[:]
```

```python
final_card_ranking = [] ##유사도 저장
count = 1
for i in data3.iloc[:, 1:len(data3)] :
    df_real = data3.iloc[:, [count]]
    df_ranking=df_real.rank(method='min',ascending=False)
    card_rank=[]
    for j in range(14) :
        card_rank.append(df_ranking.iloc[j,0])
    tau, p_value = stats.kendalltau(df_list, card_rank) ##유사도 측정
    final_card_ranking.append(tau)
    count += 1
 
```

```python
##가중치계산 준비
personal_Df_Class=personal_Df[personal_Df['카드이용금액계순위']<6].sort_values(by=personal_Df.columns[1],ascending=True).iloc[:,0]
personal_Df_Class
```

```python
#가중치이용 총점계산
card_score=['총점']
for i in range(data3.shape[1]-1) :
    score=int(data3.iloc[personal_Df_Class.index[0],i+1])*5/15 + int(data3.iloc[personal_Df_Class.index[1],i+1])*4/15 + int(data3.iloc[personal_Df_Class.index[2],i+1])*3/15 + int(data3.iloc[personal_Df_Class.index[3],i+1])*2/15 + int(data3.iloc[personal_Df_Class.index[4],i+1])*1/15
    card_score.append(score)
data2=data2.append(pd.Series(card_score,index=data2.columns),ignore_index=True)
data2
```

```python
data_score = data2.loc[17, :]
data_score
```

```python
##상위3카드 점수계산, index추출
first_card_index = 0
first_card_score = 0
second_card_index = 0
second_card_score = 0
third_card_index = 0
third_card_score = 0
change = 0
for i in range(1,len(final_card_ranking)+1):
    if final_card_ranking[i-1] >= 0.1 : #특정유사도 이상
        if float(data_score[i]) > third_card_score:
            third_card_score = float(data_score[i])
            third_card_index = i
            if float(data_score[i]) > second_card_score:
                change = second_card_score
                second_card_score = third_card_score
                third_card_score = change
                change = second_card_index
                second_card_index = third_card_index
                third_card_index = change
                if float(data_score[i]) > first_card_score:
                    change = first_card_score
                    first_card_score = second_card_score
                    second_card_score = change
                    change = first_card_index
                    first_card_index = second_card_index
                    second_card_index = change
```

```python
#상위3카드 결과출력
real_card = data2.loc[0, :]
print("1st카드명 : %s, 카드점수 : %f" %(real_card[first_card_index], data_score[first_card_index]))
print("2st카드명 : %s, 카드점수 : %f" %(real_card[second_card_index], data_score[second_card_index]))
print("3st카드명 : %s, 카드점수 : %f" %(real_card[third_card_index], data_score[third_card_index]))
```
